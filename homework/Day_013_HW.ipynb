```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

### 之前做過的處理

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

## 練習時間
參考 Day 12 範例程式，離散化你覺得有興趣的欄位，並嘗試找出有趣的訊息

```python
phone_data = app_train[['TARGET', 'DAYS_LAST_PHONE_CHANGE']].copy()
phone_data['DAYS_LAST_PHONE_CHANGE'] = abs(phone_data['DAYS_LAST_PHONE_CHANGE'])
phone_data['YEARS_LAST_PHONE_CHANGE'] = phone_data['DAYS_LAST_PHONE_CHANGE']/365
phone_data['YEARS_LAST_PHONE_CHANGE'].describe()
```

```python
phone_data['YEARS_LAST_PHONE_CHANGE'] = pd.cut(phone_data['YEARS_LAST_PHONE_CHANGE'], bins = np.linspace(0, 12, num = 7))
phone_data.head(10)
```

```python
phone_groups  = phone_data.groupby('YEARS_LAST_PHONE_CHANGE').mean()
phone_groups
```

```python
plt.figure(figsize = (8, 8))

# 繪製目標值平均與分組組別的長條圖
plt.bar(range(len(phone_groups.index)), phone_groups['TARGET'])
# 加上 X, y 座標說明, 以及圖表的標題
plt.xticks(range(len(phone_groups.index)), phone_groups.index, rotation = 75); plt.xlabel('Phone use years group'); plt.ylabel('Average Failure to Repay')
plt.title('Failure to Repay by Phone Use Year Group');
```
